# Gerando Gráficos com Matplotlib

```python
# Criaremos duas funções y1(linear) e y2(quadrática).. Criaremos um conjunto de 100 valores para x, que variam de zero a dois.
```

```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
# criar 100 dados igualmente intervalados entre zero e dois:
x = np.linspace(0,2,100)

# Função linear
y1 = x*2

# Função quadrática
y2 = x**2
```

```python
# Plotar dados
plt.plot(x,y1,'*r')
plt.plot(x,y2,'g--')

# Legenda dos eixos
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')

# Título do gráfico
plt.title('Título do gráfico')
```

```python
# Podemos customizar nossos gráficos de várias maneiras:
# modificando a cor(r- red; b- blue; yellow; g- green; dentre outros), o tipo dos pontos(*; --; *-; o; dentre outros).
# As possibilidades de formatação gráficas são praticamente infinitas, dependende de nossa criatividade.
# Consulte a documentação para mais formatos.
```

```python
# Vejamos agora uma funcionalidade bem útil:
# como obter vários gráficos um ao lado do outro. 
# Para isso, vamos gerar mais uma função chamada de y3( função cúbica).
```

```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
# Criar 100 dados igualmente intervalados
# entre zero e dois:
x = np.linspace(0,2,100)

# Função linear
y1 = x*2

# Função quadrática
y2 = x**2

# Função cúbica
y3 = x**3

# Função seno
y4 = np.sin(x)
```

```python
# Iremos utilizar a função 'subplot()'

# plots lado a lado:
# Gerar janelas com 2 linhas e com 2 colunas
fig, janela = plt.subplots(2, 2)

# Adicionar a cada janela um gráfico
janela[0,0].plot(x, y1,'r')
janela[0,1].plot(x, y2,'g')
janela[1,0].plot(x, y3,'b')
janela[1,1].plot(x, y4,'y')

fig.legend(['Linear', 'Quadrática', 'Cúbica', 'Senoidal'])
```

# Gráfico de Barras

```python
# O gráfico de barras é bastante utilizado para fazer demonstrativos estatísticos no meio 
# corporativo.
# Podemos obter um gráfico de barras com função 'bar()', na qual definimos a posição 
# das barras no qual o eixo x (localização da variáveis) 
# e sua altura no eixo y (valores da variáveis).
# De modo adicional, podemos configurar outras características outras características, como a 
# espessura das barras, cor, etc...
# representa um range com os nomes das variáveis.
# Por outro lado, o eixo Y apresenta os valores das propriedades do eixo X.
```

```python
# Vamos estudar um exemplo bem simples, onde estamos guardando od valores das vairáveis em listas.
```

```python
# Variáveis para o Bar Chart
eixo_y = [10,30,15]
eixo_x = range(len(eixo_y))

largura = 0.5
cor_barra = 'red'

plt.bar(eixo_x, eixo_y, width = largura, color = cor_barra )
plt.show()
```

```python
# Este é um exemplo básico de gráfico de barras. 
# Podemos obter gráficos muito complexos, com superposições de barras, barras laterais, 
# dentre outras.
# Ver a documentação para maiores detalhes
```

# Histogramas

```python
# Os histogramas são gráficos descritivos largamente utilizados.
# Eles geralmente são utilizados para apresentar a distribuição da frequência
# de ocorrencia de uma série de variáveis.
# A função Python nesse caso é a 'hist()'.
# Nessa função passaremos como parâmetro propriamente ditos a série de valores 
# e os #bins(são os tamanhos dos intervalos em que serão avaliadass as frequências de observação).
```

```python
# Vamos para um exemplo prático.
# Siga o seguinte raciocínio: primeiro precisamos gerar nossos dados com o  # NumPy # , 
# depois plotar o histograma com os dados.
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Gerar números aleatórios com distribuição normal
num_norm = np.random.randn(1000,1) # 1000 linhaass e 1 coluna
```

```python
# Para plotar um histograma, que é basicamente um gráfico de barras onde observamos a frequência
# de ocorrência de um determinado valor em intervalos igualmente especificados '(bins)', basta utilizar
# a função 'hist()'
```

```python
# bins = representa a quantidade de caixas com intervalos de 
# observação de frequência
plt.hist(num_norm, bins = 10)
plt.xlabel('Valores')
plt.ylabel('Frequência Observada')
plt.show()
```

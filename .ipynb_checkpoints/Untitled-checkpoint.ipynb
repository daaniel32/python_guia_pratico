```python
# importar bibliotecas para trabalhar com csv
import csv
```

```python
# Criar e salvar dados em arquivos csv
with open('DADOS.CSV', 'w') as arquivo:
    leitor =  csv.writer(arquivo)
    leitor.writerow(('Col 1', 'Col 2', 'Col 3'))
    for i in range(100):
        #salvar colunas de dados
        leitor.writerow((i,2*i, 3*i))
```

```python
# Diretório atual
os.getcwd()
```

```python
# A diretiva '..' dentro do método chdir() (change directory) nos permite retornar
# para o diretório imediatamente anterior ao atual.
# Essa troca de diretórios se dá a partir de hierarquia de diretórios mais externos
# para os mais internos e vice versa.
```

```python
# Diretório atual
os.getcwd()
```

```python
# Retornar para o diretório anterior
os.chdir('..')
```

```python
# Diretor atual
os.getcwd()
```

```python
# podemos especificar diretamente o caminho do diretório no qual queremos trabalhar.
# Algo importante é ficar atento ao sistema operacional, pois a estrutura 
# de referência de diretórios podem mudar.
# Nesse caso em específico, os exemplos deste livro estão sndo mostrado em um OS Wimsows,
# consequentemente, temos a contrabarra '\' como caractere usado para separar diretórios.
# No sistema operacional LINUX, por exemplo, temos a barra '/'.
# Vejamos agora como poodemos renomear diretórios preexistentes. Iremos renomear o diretório
# PYTHON_ESTUDO para ESTUDO_PYTHON.
```

```python
# Renomear o diretório
os.rename('ESTUDO_PYTHON','PYTHON_ESTUDO')
```

```python
os.listdir()
```

```python
# A remoção/exclusão de um diretórii e arquivos é feita a partir do método 
# .remove('nome_arquivo')
# Devemos ter a permissão de administrador do computador para fazer essa tipo de ação.
```

```python
os.remove('PYTHON_ESTUDO')
```
